Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

mu = 3
sigma = 0.5 
n=1000
vals = np.random.normal(loc=mu, scale=sigma, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=0, sigma=1)
x= np.arange(-4,4,0.001)
plt.plot(x, norm.pdf(x))
plt.title('Función de densidad normal estándar')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 1.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=3, sigma=1)
x= np.arange(0,6,0.001)
plt.plot(x, norm.pdf(x,loc=3))
plt.title('Función de densidad normal(mu=3, sigma=1)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 0.5.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=0, sigma=1)
x= np.arange(0,6,0.001)
plt.plot(x, norm.pdf(x,loc=3,scale=0.5))
plt.title('Función de densidad normal(mu=3, sigma=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

QQ plot usando statsmodels. 

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data = vals, dist=stats.norm, loc=3, scale=0.5, line="45")
plt.show()
```

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data = vals, dist=stats.norm, loc=0, scale=1, line="45")
plt.show()
```

### 3.5 Replicar Ejercicio t-Student

```python
import numpy as np

sigma = 1
n=10000
vals2 = np.random.exponential(scale=sigma, size=n)
print(vals2)
```

```python
import matplotlib.pyplot as plt


font1 = {'family': 'Segoe UI',
        'color':  'black',
        'weight': 'bold',
        'size': 14,
        }

font2 = {'family': 'Segoe UI',
        'color':  'black',
        'weight': 'semibold',
        'size': 12,
        }


count, bins, ignored = plt.hist(x=vals2, bins= round(n**(1/2)), color= '#ff4203')
plt.title('Histograma de Tiempo entre Arribos', fontdict= font1)
plt.xlabel('Tiempos entre arribos (TEA)', fontdict= font2)
plt.ylabel('Frecuencia', fontdict= font2)
plt.savefig('images/Histograma_TEA.png')
plt.show()

print(bins)
print(count)
```

### Función de Densidad Distribución Exponencial (sigma = 1)

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

# Exponencial (sigma = 0.25)
x= np.arange(0,2.5,0.001)
plt.plot(x, expon.pdf(x, scale = 1), color= '#ff4203')
plt.title(r'Función de densidad de exponencial ($\sigma$ = 1)', fontdict= font1)
plt.xlabel('Valores', fontdict= font2)
plt.ylabel('Densidad', fontdict= font2)
plt.show()
```

### Q-Q Plot Distribución Exponencial (sigma = 1)

```python
import statsmodels.api as sm
import scipy.stats as stats

fig = sm.qqplot(data = vals2, dist=stats.expon, scale=1, line="45")
plt.title(r'Q-Q plot Distribución Exponencial ($\sigma$ = 1)', fontdict= font1)
plt.xlabel('Sample Quantiles', fontdict= font2)
plt.ylabel('Theoretical Quantiles', fontdict= font2)
plt.show()
```
